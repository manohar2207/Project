```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

# Load your dataset
# Replace 'your_dataset.csv' with the actual file path
dataset = pd.read_csv('/content/COPD Dataset.csv')

# Assuming your dataset has features (X) and labels (y)
# Replace 'features' and 'labels' with the actual column names
X = dataset[['smoking', 'AGE', 'hypertension']]
y = dataset['Diabetes']

# Encode string labels to numerical values if needed
# You can use LabelEncoder as shown in the previous example

# Number of iterations
num_iterations = 20
accuracies = []

# Perform 15 iterations
for i in range(num_iterations):
    # Split the data into training and testing sets
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=i)

    # Create a Decision Tree classifier
    clf = DecisionTreeClassifier(random_state=42)

    # Train the classifier
    clf.fit(X_train, y_train)

    # Make predictions on the test set
    y_pred = clf.predict(X_test)

    # Calculate accuracy and store it
    accuracy = accuracy_score(y_test, y_pred)
    accuracies.append(accuracy)

    print(f"Iteration {i + 1}: Accuracy = {accuracy}")

# Calculate the average accuracy over all iterations
average_accuracy = np.mean(accuracies)
print(f"\nAverage Accuracy over {num_iterations} iterations: {average_accuracy}")
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder

# Load your dataset
dataset = pd.read_csv('/content/COPD Dataset.csv')

# Assuming your dataset has features (X) and labels (y)
X = dataset[['smoking', 'AGE', 'copd']]
y = dataset['Diabetes']

# Encode string labels to numerical values
le = LabelEncoder()
y = le.fit_transform(y)

# Number of iterations
num_iterations = 20
accuracies = []

# Perform 15 iterations
for i in range(num_iterations):
    # Split the data into training and testing sets
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=i)

    # Create a Random Forest classifier
    clf = RandomForestClassifier(n_estimators=100, random_state=42)

    # Train the classifier
    clf.fit(X_train, y_train)

    # Make predictions on the test set
    y_pred = clf.predict(X_test)

    # Calculate accuracy and store it
    accuracy = accuracy_score(y_test, y_pred)
    accuracies.append(accuracy)

    print(f"Iteration {i + 1}: Accuracy = {accuracy}")

# Calculate the average accuracy over all iterations
average_accuracy = np.mean(accuracies)
print(f"\nAverage Accuracy over {num_iterations} iterations: {average_accuracy}")
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

# Load your dataset
# Replace 'your_dataset.csv' with the actual file path
dataset = pd.read_csv('/content/COPD Dataset.csv')

# Assuming your dataset has features (X) and labels (y)
# Replace 'features' and 'COPD_label' with the actual column names
X = dataset[['AGE', 'smoking', 'copd']]
y = dataset['Diabetes']

# Encode string labels to numerical values if needed
# You can use LabelEncoder as shown in previous examples

# Number of iterations
num_iterations = 20
accuracies = []

# Perform 20 iterations
for i in range(num_iterations):
    # Split the data into training and testing sets
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=i)

    # Create an SVM classifier
    svm_classifier = SVC(kernel='linear', random_state=42)

    # Train the classifier
    svm_classifier.fit(X_train, y_train)

    # Make predictions on the test set
    y_pred = svm_classifier.predict(X_test)

    # Calculate accuracy and store it
    accuracy = accuracy_score(y_test, y_pred)
    accuracies.append(accuracy)

    print(f"Iteration {i + 1}: Accuracy = {accuracy}")

# Calculate the average accuracy over all iterations
average_accuracy = np.mean(accuracies)
print(f"\nAverage Accuracy over {num_iterations} iterations: {average_accuracy}")
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# Load your dataset
# Replace 'your_dataset.csv' with the actual file path
dataset = pd.read_csv('/content/COPD Dataset.csv')

# Assuming your dataset has features (X) and labels (y)
# Replace 'features' and 'COPD_label' with the actual column names
X = dataset[['AGE', 'smoking', 'copd']]
y = dataset['Diabetes']

# Encode string labels to numerical values if needed
# You can use LabelEncoder as shown in previous examples

# Number of iterations
num_iterations = 20
accuracies = []

# Perform 20 iterations
for i in range(num_iterations):
    # Split the data into training and testing sets
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=i)

    # Create a KNN classifier
    k_neighbors = 5  # You can adjust the number of neighbors
    knn_classifier = KNeighborsClassifier(n_neighbors=k_neighbors)

    # Train the classifier
    knn_classifier.fit(X_train, y_train)

    # Make predictions on the test set
    y_pred = knn_classifier.predict(X_test)

    # Calculate accuracy and store it
    accuracy = accuracy_score(y_test, y_pred)
    accuracies.append(accuracy)

    print(f"Iteration {i + 1}: Accuracy = {accuracy}")

# Calculate the average accuracy over all iterations
average_accuracy = np.mean(accuracies)
print(f"\nAverage Accuracy over {num_iterations} iterations: {average_accuracy}")
```
